```python
from selenium import webdriver
from selenium.webdriver.edge.service import Service
from selenium.webdriver.common.by import By
import pandas as pd
from tqdm import tqdm
from lst_py import lst_urls

# Especifica la ruta del WebDriver
webdriver_path = r'.\edgedriver_win64\msedgedriver.exe'

# Crea un objeto Service y el WebDriver una vez
service = Service(webdriver_path)
driver = webdriver.Edge(service=service)

lst_global = []

for url in tqdm(lst_urls):

    driver.get(url)
    driver.implicitly_wait(1) 

    url_data = {}
    url_data["url"] = url

    try:
        # fecha
        tbl_date = driver.find_element(By.CSS_SELECTOR, 'table.table.table_summary')
        tbody = tbl_date.find_element(By.TAG_NAME, "tbody")
        rows = tbody.find_elements(By.TAG_NAME, "tr")

        for ir in range(len(rows)):
            key = "vigencia_years" if ir == 0 else rows[ir].find_element(By.TAG_NAME, 'td').get_attribute('innerText').replace(" ", "_").strip()
            val = rows[ir].find_element(By.TAG_NAME, 'td').get_attribute('innerText') if ir == 0 else rows[ir].find_elements(By.TAG_NAME, 'td')[1].get_attribute('innerText').strip()
            val = "SIN INFORMACION" if len(val) == 0 else val
            url_data[key] = val

        # DATOS EDO MUN POB
        data_eml = driver.find_element(By.ID, "tblTvFrecuenciaPrincipal")
        tbl_data_head = data_eml.find_element(By.TAG_NAME, "thead").find_elements(By.TAG_NAME, "th")
        tbl_data_tbody = data_eml.find_element(By.TAG_NAME, "tbody").find_elements(By.TAG_NAME, "td")

        for i in range(len(tbl_data_head)):
            key = tbl_data_head[i].get_attribute("innerText").strip().replace(" ", "_")
            value = tbl_data_tbody[i].get_attribute("innerText").strip()
            value = "SIN INFORMACIÓN" if len(value) == 0 else value
            url_data[key] = value

        lst_global.append(url_data)
    
    except Exception as e:
        print(f"Error processing {url}: {e}")

driver.quit()

# Opcional: Guarda los datos en un archivo
df = pd.DataFrame(lst_global)
df.to_csv('output.csv', index=False)
```
